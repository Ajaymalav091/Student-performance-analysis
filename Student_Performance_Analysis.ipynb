```python
ax = sns.countplot(x='gender' ,data=df ) # Comparsion accoding Gender 
for i in ax.containers:
    ax.bar_label(i)
```

```python
df.groupby(['gender'],as_index=False)['math_score'].sum().sort_values(by='math_score',ascending=False)
```

### Results : From above graph we see that the most of the student are females passed highest scores. 

```python
# Comparsion accoding parental_level_of_education
sns.set(rc={'figure.figsize':(15,5)}) 
ax = sns.countplot(x='parental_level_of_education',data=df)
for i in ax.containers:
    ax.bar_label(i)
```

```python
df.groupby(['parental_level_of_education'],as_index=False)['math_score'].sum().sort_values(by='math_score',ascending=False)
```

### Result : From above graph we see that the most of the student are  collage students passed highest scores.
